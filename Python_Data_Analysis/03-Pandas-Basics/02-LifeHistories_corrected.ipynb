## Step 3: Find a simple model for the data

Using for instance [scipy's linear regression function](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.linregress.html), find and plot a simple model between the adults' and the offsprings' mass. Plot the resulting model over the data, label the axes and add a legend for both the experimental data and the resulting model. 

**Hint**: You will need to deal with missing values.

By applying a linear regression on the log-log data, we obtain:

$ log(adult) = intercept + slope \times log(newborn)$

Which can be rewritten as:

$ adult = e^{intercept} \times newborn^{slope}$

```python
# Apply a linear regression to the log-log data
from scipy import stats
# Grab the initial data
adult_mass = df["mass(g)"].values
newborn_mass = df["newborn(g)"].values
# Remove the points for which either the adult's or the newborn's mass is nan
to_delete = np.isnan(adult_mass) | np.isnan(newborn_mass)
adult_mass = adult_mass[~to_delete]
newborn_mass = newborn_mass[~to_delete]

slope, intercept, r_value, p_value, std_err = stats.linregress(np.log(adult_mass), np.log(newborn_mass))
```

```python
print(f"""------OUTPUT OF THE REGRESSION------
Slope: {slope}
Intercept: {intercept}""")
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

# Compute linearly spaced points on a log scale
x_regression = np.logspace(0, 8, 100)
# Resulting regression using the previously computed coefficients
y_regression = np.exp(intercept) * x_regression**slope

ax.scatter(df["mass(g)"], df["newborn(g)"], s=40, label="Experimental data")
ax.plot(x_regression, y_regression, c="k", label="Regression")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("log mass")
ax.set_ylabel("log newborn mass")
ax.set_title("Adult vs Newborn log mass")
ax.legend()
```
